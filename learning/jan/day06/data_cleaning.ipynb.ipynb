```python
import pandas as pd
```

```python
dataset = pd.read_csv("Student.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset["Age"].mode()[0]
```

```python
dataset["Marks"] = dataset["Marks"].fillna(dataset["Marks"].mean())
dataset["Age"] = dataset["Age"].fillna(dataset["Age"].mode()[0])
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isnull().sum()
```

```python
filtered = dataset[dataset["Marks"]>74]
print("above 74 marks Students\n",filtered)
```
