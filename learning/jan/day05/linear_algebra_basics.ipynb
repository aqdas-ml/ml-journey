import NumPy module

```python
import numpy as np
```

Create Vectors

```python
v1 = np.array([4,5,6])
v2 = np.array([2,3,4])
```

```python
m1 = np.array([
    [4,5],
    [6,7]
])
m2 = np.array([
    [2,3],
    [1,8]
])
```

Dot Product

```python
print("Dot product of Vectors: ")
np.dot(v1,v2)
```

The dot product combines corresponding elements of both vectors and sums them.
This result represents how input values are weighted to produce a single output.

Matrix Multiplication

```python
print("Matrix Multiplication: ")
np.dot(m1,m2)
```

Matrix multiplication applies dot products between rows and columns.
The resulting matrix contains combined values from both input matrices.
