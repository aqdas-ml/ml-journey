```python
import pandas as pd
dataset = pd.read_csv("students_multi.csv")
dataset.head()
```

```python
dataset["Study_efficiency"] = dataset["Study_hours"] / dataset["Attendance"]
dataset.head()
```

A new feature, Study efficiency, is created by combining Study hours and Attendance.

This feature may capture student effort more effectively.


```python
x = dataset[["Age", "Study_hours","Attendance", "Study_efficiency"]]
y = dataset["Marks"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=42)

model = LinearRegression()
model.fit(x_train,y_train)

y_pred = model.predict(x_test)
print("r2_score: ", r2_score(y_test,y_pred))
```

The engineered feature improves model performance by providing additional meaningful information.

Feature engineering can improve results more effectively than changing models.


# Conclusion:

Feature engineering and scaling significantly impact model performance.

Improving features is often more beneficial than switching to complex models.
