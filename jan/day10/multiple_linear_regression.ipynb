```python
import pandas as pd
dataset = pd.read_csv("students_multi.csv")
dataset.head()
```

This dataset contains multiple features (Age, Study hours, Attendance) used to predict Marks.

Multiple linear regression is suitable when several linear features influence the target.


```python
import matplotlib.pyplot as plt

plt.scatter(dataset["Study_hours"], dataset["Marks"])
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.show()
```

The scatter plot shows a positive linear relationship between Study hours and Marks.

This indicates that Study hours is a strong predictor.


```python
x = dataset.iloc[:,:-1]
y = dataset["Marks"]
```

Multiple input features are used to predict Marks, which is the target variable.

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

The dataset is split into training and testing sets to evaluate model generalization.


```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = model.predict(x_test)

print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

The model achieves a high R² score, indicating that multiple features together explain most of the variance in Marks.


```python
coeff_ds = pd.DataFrame({
    "Feature": x.columns,
    "Coefficient": model.coef_
})

coeff_ds
```

The coefficients show the contribution of each feature to the prediction.
    
Higher coefficients indicate stronger influence on Marks.


Conclusion:

Multiple linear regression performs well because the relationship between features and target is linear.
Using multiple relevant features improves prediction compared to single-feature linear regression.
