```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Student.csv")
dataset.head()
```

```python
dataset.describe()
```

```python
dataset["Age"] = dataset["Age"].fillna(dataset["Age"].mode()[0])
dataset["Marks"] = dataset["Marks"].fillna(dataset["Marks"].mean())
```

```python
dataset.isnull().sum()
```

```python
x = dataset["Marks"]
y = dataset["Age"]
plt.scatter(x,y)
plt.xlabel("Marks")
plt.ylabel("Age")
plt.title("Age and Marks Graph")
plt.show()
plt.savefig("scatter_plot.png")
```

This scatter plot shows the relationship between students' age and their marks.
There is no clear linear trend, indicating that age does not strongly affect marks in this dataset.
