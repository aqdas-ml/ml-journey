```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Student.csv")
dataset.head()
```

```python
dataset.describe()
```

```python
dataset["Age"] = dataset["Age"].fillna(dataset["Age"].mode()[0])
dataset["Marks"] = dataset["Marks"].fillna(dataset["Marks"].mean())
```

```python
data = dataset["Marks"]
plt.hist(data,bins=len(data))
plt.show()
```

This histogram shows the distribution of student marks.
Most students have marks clustered in the higher range, indicating generally good performance.

```python
x = dataset["Name"]
y = dataset["Marks"]
plt.figure(figsize=(10,5))
plt.bar(x,y)
plt.show()
```

This bar chart compares marks scored by each student.
It helps visually identify which student has higher or lower marks.
